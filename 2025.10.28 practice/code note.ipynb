## K-Means++ Initialization code defination

```python
import matplotlib.pyplot as plt

from sklearn.cluster import kmeans_plusplus
from sklearn.datasets import make_blobs
```

### 导入所需的库
* matplotlib.pyplot 用于绘制数据点与聚类结果的可视化图像
* kmeans_plusplus kmeans聚类算法 函数生成聚类初始种子的输出，用于K-Means  初始化算法
* make_blobs 用于生成聚类用的模拟数据

```python
n_samples = 4000
n_components = 4 # 指定数据集有四个簇
```

```python
X, y_true = make_blobs(
    n_samples=n_samples,       # 样本总数
    centers=n_components,      # 簇数
    cluster_std=0.60,          # 每个簇的标准差（决定簇的紧密程度）
    random_state=0             # 固定随机种子，保证结果可重复
)
```

```python
X = X[:, ::-1]  #将两个维度顺序反转
```

```python
centers_init, indices = kmeans_plusplus(
    X,
    n_clusters = 4,
    random_state = 0 # 固定随机种子以确保结果一致
)
```

```python
plt.figure(1)
colors = ["#4EACC5", "#FF9C34", "#4E9A06", "m"]
for k, col in enumerate(colors):
    cluster_data = y_true == k         # 找出属于第 k 个簇的样本点
    plt.scatter(
        X[cluster_data, 0],            # 取出该簇的第1维坐标
        X[cluster_data, 1],            # 取出该簇的第2维坐标
        c=col,                         # 颜色
        marker=".",                    # 使用小圆点
        s=10                           # 点的大小
    )
# 绘制初始化得到的聚类中心，用蓝色、较大点标示
plt.scatter(centers_init[:, 0], centers_init[:, 1], c="b", s=50)

# 设置图表标题
plt.title("K-Means++ Initialization")

# 去掉坐标轴刻度，保持视觉简洁
plt.xticks([])
plt.yticks([])

# 显示绘制结果
plt.show()
```
